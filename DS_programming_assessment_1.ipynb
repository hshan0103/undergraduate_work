```python
# Stage 1: Reading and writing data to and from files
def read_from_file(filename):
    # Enter your code here
    with open(filename, 'r') as file:
        lines = file.read()
        lines = lines.split()
        data = []
        for line in lines:
            if line != '-':
                data.append(round(float(line),2))
    return (data)

def write_to_file(filename, binned_data):
    # Enter your code here
    with open(filename, 'w') as file:
        write_line = ''
        for index, count_point in binned_data.items():
            write_line = write_line + str(index) + ': ' + str(count_point) + '\n'
        write_line = write_line[:-1]
        file.writelines(write_line)
```

```python
# Stage 2: Place the data into bins
def bin_data(data, binsize):
    # Enter your code here
    lower_bound = min(data)
    max_data = max(data)
    
    dict = {}
    
    while lower_bound <= max_data:
        #setting the lower bound(low) and upper bound(high) for each bin
        low = round(float(lower_bound),2)
        high = round(float(low + binsize),2)
        bin_intervals = (str(low) + ' - ' + str(high))
        
        count_data = 0
        #looping the dataset, count the data falls in particular intervals of bins
        for point in data:
            if point >= low and point < high:
                count_data += 1
        dict[bin_intervals] = count_data
        lower_bound = high
        
    return (dict)
    

# Stage 3: Display the frequency analysis
def display_freq_analysis(binned_data):
    # Enter your code here
    for index, count_point in binned_data.items():
        print(str(index), ': ', str(count_point), sep='')
```

```python

# Stages 4 and 5: Interact with the user with menus and error handling.
# The following line suppresses the code when the file is imported 

#create class for main menu
class Menus:
        def __init__(self):
            self.options = '1. Set bin size' + '\n' + '2. Display frequency analysis' + '\n' + '3. Exit'
            print(self.options)
            self.choose = input('Please enter your choice: ')
        def choice(self):
            return (self.choose)
    
if __name__ == '__main__':

    # Enter your code here (indented)
    while True:
        filename = input('Enter data file: ')
        try:
            #stop looping if valid name input
            data = read_from_file(filename)
            break
        except FileNotFoundError:
            #keep looping until valid filename input
            print('File not available')
            continue

    menu = Menus()    
    binsize = ''
    
    while menu.choice() != '3':
    #possible invalid selection before exiting, while loop inside while selection != 3
        while menu.choice() not in ['1','2','3']:
            print('Not a valid option')
            menu = Menus()
            
        #option 1: set bin size
        if menu.choice() == '1':
            binsize = float(input('Enter bin size: '))
            menu = Menus()
        elif menu.choice() == '2':
            
            #option 2: display analysis
            #if no record of bin size input since starting the program, back to main menu
            if binsize == '':
                print('Please set a bin size first')
                menu = Menus()
                
            #creating analysis according to the last bin size input    
            else:
                binned_data = bin_data(data, binsize)
                display_freq_analysis(binned_data)
                save_file = input('Save analysis to file? (Y/N) ')
                while save_file not in ['Y','N']:
                    print('Not a valid option')
                    save_file = input('Save analysis to file? (Y/N) ')
                if save_file == 'Y':
                    w_filename = input('Enter file name: ')
                    write_to_file(w_filename, binned_data)
                    menu = Menus()
                else:
                    menu = Menus()
    else:
        #option 3: exit program
        exit()

```
